Bắt đầu


1.1.1. Quy trình khai thác dữ liệu và ra quyết định dựa trên dữ liệu (insight-driven methodology). 

1. Data Collection (Thu thập dữ liệu)
2. Data Processing (Xử lý dữ liệu)
    Làm sạch dữ liệu, chuẩn hóa, loại bỏ lỗi, format lại,...
3. Data Visualization & Grasping (Trực quan hóa và hiểu dữ liệu)
    Dùng biểu đồ, bảng, heatmap,... để "nhìn" được xu hướng và hiểu bản chất của dữ liệu.
4. Analysis, Hypothesis Testing & ML (Phân tích, kiểm định giả thuyết và học máy)
5. Insight & Policy Decision (Rút ra insight và đưa ra quyết định)
    Từ kết quả phân tích, đưa ra gợi ý hành động cụ thể cho doanh nghiệp hoặc tổ chức.

1.1.2. Phương pháp luận phân tích dữ liệu theo hướng sản phẩm (product-driven) 

1. Business understanding (Hiểu biết nghiệp vụ)
2. Analytic approach (Phương pháp tiếp cận)
3. Data requirements (Yêu cầu dữ liệu)
4. Data collection (Thu thập dữ liệu)
5. Data understanding (Hiểu dữ liệu)
6. Data preparation (Chuẩn bị dữ liệu)
7. Modeling (Xây dựng mô hình)
8. Evaluation (Đánh giá mô hình)
9. Deployment (Triển khai mô hình)
10. Feedback (Phản hồi)
